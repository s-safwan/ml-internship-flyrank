## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

**Why ML beats a fixed rule**: A fixed human rule (like "pages older than 180 days with low CTR") creates massive blocks of tied scores. If a simple rule flags 5,000 pages, the editorial team still doesn't know which one to fix first. Machine Learning solves this by weighing multiple signals simultaneously (age, impressions, position, engagement) to break those ties and produce a sharp, continuous 1-to-N ranking.
